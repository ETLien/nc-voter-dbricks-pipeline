```python
import os
import z_pipelineUtils as z_pU

# Create directories
VOLUME = "/Volumes/nc_pipeline/default/pipeline_data"
zip_dir = f"{VOLUME}/zipped"
unzip_dir = f"{VOLUME}/unzipped"
os.makedirs(VOLUME, exist_ok=True)
os.makedirs(zip_dir, exist_ok=True)
os.makedirs(unzip_dir, exist_ok=True)

# Create the pipeline_state.json file if it doesn't already exist
# The intent for this file is to store information about the latest run of the pipeline
# This will be referenced on subsequent runs of the pipeline to determine if the full pipeline should be run
# The structure is a kv-pair, where the key is the zipfile and it's value is a nested dictionary
# 'url' represents the URL where the file is available for download on the S3
# 'etag' represents a unique string that can be used to determine if a new file has been uploaded to the S3
# 'content_length' represents the filesize of the download; will be used to calculate download completion
# 'downloaded' represents if the new file has been downloaded
# 'unzipped' represents if the new file has been unzipped after download
# 'tested' represents if the new file has meets certain criteria and passes initial integrity checks
# 'archived' represents if the target table in the database has been backed up prior to any changes from the current run of the pipeline are performed
# 'processed' represents if the table in the database has been updated with the latest changes from the current run of the pipeline
json_file = f"{VOLUME}/pipeline_state.json"
if not os.path.exists(json_file):
    default_state = {
        "ncvoter_Statewide.zip": {
            "url": "https://s3.amazonaws.com/dl.ncsbe.gov/data/ncvoter_Statewide.zip",
            "etag": "",
            "content_length": "",
            "downloaded": False,
            "unzipped": False,
            "tested": False,
            "archived": False,
            "processed": False,
        },
        "ncvhis_Statewide.zip": {
            "url": "https://s3.amazonaws.com/dl.ncsbe.gov/data/ncvhis_Statewide.zip",
            "etag": "",
            "content_length": "",
            "downloaded": False,
            "unzipped": False,
            "tested": False,
            "archived": False,
            "processed": False,
        }
    }
    z_pU.updateStateJson(default_state, json_file)
    print(f"Created pipeline_state.json file at {json_file}")
else:
    print(f"pipeline_state.json file already exists at {json_file}")

```
