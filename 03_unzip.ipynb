```python
import os
import zipfile
import z_pipelineUtils as z_pU

# File that stores the most recent run details of the pipeline
PIPELINE_STATE = "/Volumes/nc_pipeline/default/pipeline_data/pipeline_state.json"
# Directory that will store the downloaded zip files
DOWNLOAD_DEST = "/Volumes/nc_pipeline/default/pipeline_data/zipped"
# Directory that will store the unzipped files
UNZIP_DEST = "/Volumes/nc_pipeline/default/pipeline_data/unzipped"

saved_state, live_state = z_pU.readStateJson(PIPELINE_STATE)

for voter_file in saved_state:
    unzipped = saved_state[voter_file]["unzipped"]
    full_zip_path = os.path.join(DOWNLOAD_DEST, voter_file)
    if (downloaded is True) and (unzipped is False):
        with zipfile.ZipFile(full_zip_path, "r") as zip:
            zip.extractall(UNZIP_DEST)
        print(f"{voter_file} has been unzipped to {UNZIP_DEST}.")
        live_state[voter_file]["unzipped"] = True
        z_pU.updateStateJson(live_state, PIPELINE_STATE)
    elif unzipped is False:
        print(f"{voter_file} has not been downloaded yet, skipping unzip.")
    else:
        print(f"{voter_file} was already unzipped.")

```
